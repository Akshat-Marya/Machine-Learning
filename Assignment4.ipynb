```python
import numpy as np
import matplotlib.pyplot as plt

file = np.loadtxt("C:\\Users\\aksha\\PycharmProjects\\FirstProject\\venv\\pattern1.txt")

final_list = np.array_split(file,26)

############################## >>>Multi Layer Perceptron<<< ################################################



X = final_list

for i in range(len(X)):
    X[i] = np.array(X[i]).ravel()

for i in range(len(X[0])):
    X[0][i] = np.array(X[0][i]) - np.array(X[0][i])


Y = np.array([1, 0, 0, 0, 0, 0, 1])

for i in range(25):
    temp = np.array(list(map(int, list('{0:b}'.format(66 + i)))))
    Y = np.vstack((Y, temp))
    # np.array(list(map(int, list(Y))))

# model specification
Ni = 156;
Nh = 26;
No = 7;

# paramteres and array intialisation
Ntrials = 1000
wh = np.random.randn(Nh, Ni);
dwh = np.zeros(wh.shape)
wo = np.random.randn(No, Nh);
dwo = np.zeros(wo.shape)
error = np.array([])


# np.array(X).ravel()

# for i in range(len(X)):
#    X[i]=np.array(X[i]).ravel()


for trials in range(Ntrials):
    h = 1 / (1 + np.exp(-wh @ np.array(X).T))  # hidden activation for all pattern
    y = 1 / (1 + np.exp(-wo @ h))  # output for all pattern
    do = y * (1 - y) * (Y.T - y)
    dh = h * (1 - h) * (wo.transpose() @ do)  # delta backpropogated

    # update weights with momentum
    dwo = 0.9 * dwo + do @ h.T
    wo = wo + 0.1 * dwo
    dwh = 0.9 * dwh + dh @ np.array(X)
    wh = wh + 0.1 * dwh
    error = np.append(error, np.sum(abs(Y.T - y) ** 2))
plt.plot(error)
plt.xlabel("Iteration")
plt.ylabel("Error")
plt.show()

y1=y.T
error1=error


#convert the features of each character to its nearest round number.
y = np.around(y,decimals=1).T

#convert each feature set into a string for each character
l=[]
for i in y:
    l = ['-'.join(str(int(x))) for x in i]
    l = np.array2string(i.astype(int),separator='')
    l = l.strip("[]")
    #print(l)
    print(chr(int(l,2)))
```
